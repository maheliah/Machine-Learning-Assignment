# Min Max Scaler (Normalization)

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max_scaler=MinMaxScaler()
```

```python
df2=pd.read_csv('agora.csv')
```

```python
x2=df2.drop('Profit',axis=1)
```

```python
x2.head(6)
```

```python
df2['Marketing Spend']=min_max_scaler.fit_transform(x2[['Marketing Spend']])
df2['Administration']=min_max_scaler.fit_transform(x2[['Administration']])
df2['Transport']=min_max_scaler.fit_transform(x2[['Transport']])
```

```python
df2.head(5)
```
